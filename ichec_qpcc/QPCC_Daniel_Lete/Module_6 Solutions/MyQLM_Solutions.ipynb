# First Bell State

# 1/sqrt(2) * ( |00⟩ + |11⟩ )

```python
# Create a circuit to generate a Bell State:
# 1/sqrt(2) * ( |00⟩ + |11⟩ )

qprog_1 = Program()
nqbits_1 = 2 
qbits_1 = qprog_1.qalloc(nqbits_1)
# Apply some quantum Gates
qprog_1.apply(H, qbits_1[0]) 
qprog_1.apply(CNOT, qbits_1[0], qbits_1[1])

# Export this program into a quantum circuit
circuit_1 = qprog_1.to_circ()
# And display it!

plot_in_notebook(circuit_1)
```

```python

# Create a Quantum Processor Unit
pylinalgqpu = PyLinalg()

# Create a job
job_1 = circuit_1.to_job()

# Submit the job to the QPU
result_1 = pylinalgqpu.submit(job_1)

# Iterate over the final state vector to get all final components
for sample in result_1:
    print("State %s amplitude %s Probability %s" % (sample.state, sample.amplitude, sample.probability))
print('\n')
# Create a job where we specify the number of shots
job_1_sample = circuit_1.to_job(nbshots=1024)

# Submit to the QPU
result_1_sample = pylinalgqpu.submit(job_1_sample)

# And use it to draw 10 samples
for sample in result_1_sample:
    print("We measured the state {} (its probabability computed for 1024 shots is {})".format(sample.state, sample.probability))
```

```python
x = [result_1_sample[0].state, result_1_sample[1].state]
default_x_ticks = range(len(x))
plt.xticks(default_x_ticks, x)
plt.bar([0,1],[result_1_sample[0].probability, result_1_sample[1].probability], width=0.25)
plt.xlabel('state')
plt.ylabel('probability')
plt.show()
```
